```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Test Case 1: Using constant density

```python
import SPHA
```

```python
n = 101
x = np.mgrid[-1:1:n*1j]
```

```python
dx = (max(x) - min(x))/(n-1)
```

```python
h = dx
```

```python
f = np.sin(2*np.pi*x)
```

```python
y = SPHA.sph_approx(x, f, h, kernel = SPHA.spline)
```

```python
plt.plot(x,y,'o')
plt.legend('Spline')
plt.show()
```

```python
y = SPHA.sph_approx(x, f, h, kernel = SPHA.gauss)
plt.plot(x,y,'o')
plt.legend('Gauss')
plt.show()
```

Now lets look at the approximation of the derivative of the sine function.

```python
y = SPHA.sph_approx(x, f, h, kernel = SPHA.gaussd)
plt.plot(x,y,'o')
plt.legend('Gauss')
plt.show()
```

```python
y = SPHA.sph_approx(x, f, h, kernel = SPHA.splined)
plt.plot(x,y,'o')
plt.legend('Spline')
plt.show()
```

# Test Case 2: Density as a kernel approximation

```python
m = 0.02
y = SPHA.sph_approx_density(x, f, h, m, kernel = SPHA.gauss)
plt.plot(x,y,'o')
plt.legend('Gauss')
plt.show()
```

```python
y = SPHA.sph_approx_density(x, f, h, m, kernel = SPHA.spline)
plt.plot(x,y,'o')
plt.legend('Spline')
plt.show()
```

```python
y = SPHA.sph_approx_density_derivative(x, f, h, m, kernel = SPHA.spline, kerneld = SPHA.splined)
plt.plot(x,y,'o')
plt.legend('Spline')
plt.show()
```

```python
y = SPHA.sph_approx_density_derivative(x, f, h, m, kernel = SPHA.gauss, kerneld= SPHA.gaussd)
plt.plot(x,y,'o')
plt.legend('Gauss')
plt.show()
```

1. We can see that both the cases gives identical results. So we can approximate a function either by using constant density or by using density approximation.

2. The difference can be seen when we plot the error (L2 error) for varoius 'h', which we will see in the next notebook.
