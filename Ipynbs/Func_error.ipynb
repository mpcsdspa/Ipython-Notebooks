# Test Case 1: Keeping density constant

```python
n = 101
x = np.mgrid[-1:1:n*1j]
m = 0.02
```

```python
spline.error(x)
spline.l2error(x)
```

```python
gauss.error(x)
gauss.l2error(x)
```

```python
spline_derivative.error(x)
spline_derivative.l2error(x)
```

```python
gauss_derivative.error(x)
gauss_derivative.l2error(x)
```

We can clearly see from the above plots as we vary h, there exists an optimum value of h/dx where the L2 error is minimum. From the plots the optimum value of h/dx lies in the region [0.9, 1.1]
