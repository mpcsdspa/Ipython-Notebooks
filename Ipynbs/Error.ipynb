### Density

```python
import matplotlib.pyplot as plt
import numpy
```

```python
cd ~/Desktop/try/error/txt_files/density/
```

```python
%run code.py
```

We can see from the above plots that the error obtained for the density profile decreases as the number of particles(n) is increased. The solution converges to the exact Riemann solution. Now, on plotting on log-log scale, we get a linear decrease with a slope of -1.86 (~2). From this observation, we can say that the error converges in the order of O($n^2$)

```python
x = []
r1 = []
r2 = []
r3 = []
r4 = []
with open('aaa.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(float(row[0]))
        r1.append(float(row[2]))
        r2.append(float(row[3]))
        r3.append(float(row[4]))
        r4.append(float(row[5]))
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.plot(x, r1, marker='o')
plt.xlabel('Number of particles')
plt.ylabel('L1 Error')
plt.title('Region-1')
plt.subplot(222)
plt.plot(x, r2, marker='o')
plt.xlabel('Number of particles')
plt.ylabel('L1 Error')
plt.title('Region-2')
plt.subplot(223)
plt.plot(x, r3, marker='o')
plt.xlabel('Number of particles')
plt.ylabel('L1 Error')
plt.title('Region-3')
plt.subplot(224)
plt.plot(x, r4, marker='o')
plt.xlabel('Number of particles')
plt.ylabel('L1 Error')
plt.title('Region-4')
plt.show()
```

```python
x = []
y = []
with open('aaa.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[0])
        y.append(row[6])
plt.plot(x,y,marker='o')
plt.xlabel('Number of particles')
plt.ylabel('L_inf Error')
plt.title('L_inf error vs n')
plt.show()
```

However from the above plot, we can infer that the infinite norm error for the density profile increases as the number of particles are increased(except in some cases). And also the maximum error is always obtained at the location of discontinuity (at x = 0.25).
